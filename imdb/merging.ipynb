```python
import glob
import pandas as pd
```

```python
all_files = glob.glob("imdb_batch_*.csv")
df_final = pd.concat([pd.read_csv(f) for f in all_files], ignore_index=True)
df_final.to_csv("imdb_all.csv", index=False)
```

```python
df_box = df_final[df_final['Box_Office'].notna()].copy()
```

```python
df_box.head()
```

```python
#Extracting the digits for Box Office
df_box['box_value'] = df_box['Box_Office'].str.extract(r'(\d+\.?\d*)')
df_box['bud_value'] = df_box['Budget'].str.extract(r'(\d+\.?\d*)')
```

```python
df_box
```

```python
#Convert to float
df_box['box_value'] = df_box['box_value'].astype(float)
df_box['bud_value'] = df_box['bud_value'].astype(float)

#Multiplying for million and crores
df_box['multiplier'] = 1
df_box.loc[df_box['Box_Office'].str.contains(r'crore[s]?', case=False, na=False), 'multiplier'] = 1e7
df_box.loc[df_box['Box_Office'].str.contains(r'million[s]?', case=False, na=False), 'multiplier'] = 1e6
df_box.loc[df_box['Box_Office'].str.contains(r'lakh[s]?', case=False, na=False), 'multiplier'] = 1e5
df_box.loc[df_box['Box_Office'].str.contains(r'billion[s]?', case=False, na=False), 'multiplier'] = 1e9
df_box.loc[df_box['Box_Office'].str.contains('cr', case=False, na=False), 'multiplier'] = 1e7
df_box['bo_num'] = (df_box['box_value'] * df_box['multiplier']).round(0)

df_box['bud_multiplier'] = 1
df_box.loc[df_box['Budget'].str.contains(r'crore[s]?', case=False, na=False), 'bud_multiplier'] = 1e7
df_box.loc[df_box['Budget'].str.contains(r'million[s]?', case=False, na=False), 'bud_multiplier'] = 1e6
df_box.loc[df_box['Budget'].str.contains(r'lakh[s]?', case=False, na=False), 'bud_multiplier'] = 1e5
df_box.loc[df_box['Budget'].str.contains(r'billion[s]?', case=False, na=False), 'bud_multiplier'] = 1e9
df_box.loc[df_box['Budget'].str.contains('cr', case=False, na=False), 'bud_multiplier'] = 1e7
df_box['bud_num'] = (df_box['bud_value'] * df_box['bud_multiplier']).round(0)



```

```python
df_box.to_csv("check3.csv", index=False)
```
